```python
import sqlite3
import os

def connect():
    conn = sqlite3.connect("bill.db")
    cur = conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS bill (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, price REAL, qty INTEGER)")
    conn.commit()
    conn.close()

def add_item():
    name = input("Enter item name: ")
    price = float(input("Enter price: "))
    qty = int(input("Enter quantity: "))
    conn = sqlite3.connect("bill.db")
    cur = conn.cursor()
    cur.execute("INSERT INTO bill (name, price, qty) VALUES (?, ?, ?)", (name, price, qty))
    conn.commit()
    conn.close()

def view_bill():
    conn = sqlite3.connect("bill.db")
    cur = conn.cursor()
    cur.execute("SELECT * FROM bill")
    rows = cur.fetchall()
    conn.close()
    if not rows:
        print("Bill is empty")
        return
    total = 0
    print("{:<5} {:<20} {:<10} {:<10} {:<10}".format("ID", "Item", "Price", "Qty", "Total"))
    for row in rows:
        item_total = row[2] * row[3]
        total += item_total
        print("{:<5} {:<20} {:<10.2f} {:<10} {:<10.2f}".format(row[0], row[1], row[2], row[3], item_total))
    print(f"\nTotal Amount: {total:.2f}")

def remove_item():
    item_id = input("Enter item ID to remove: ")
    conn = sqlite3.connect("bill.db")
    cur = conn.cursor()
    cur.execute("DELETE FROM bill WHERE id = ?", (item_id,))
    conn.commit()
    conn.close()

def clear_bill():
    conn = sqlite3.connect("bill.db")
    cur = conn.cursor()
    cur.execute("DELETE FROM bill")
    conn.commit()
    conn.close()

def display_menu():
    print("1. Add Item")
    print("2. View Bill")
    print("3. Remove Item")
    print("4. Clear Bill")
    print("5. Exit")

def clear():
    os.system('cls' if os.name == 'nt' else 'clear')

def success():
    connect()
    while True:
        clear()
        display_menu()
        choice = input("Enter your choice: ")
        if choice == '1':
            add_item()
        elif choice == '2':
            view_bill()
        elif choice == '3':
            remove_item()
        elif choice == '4':
            clear_bill()
        elif choice == '5':
            break
        else:
            print("Invalid choice")
        input("\nPress Enter to continue...")

success()
```
